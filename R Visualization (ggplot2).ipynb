```r
# install.packages("ggplot2")
# Load the ggplot2 package
library(ggplot2)
```

```r
# Create a data frame
df <- data.frame(
    Name = c("a", "b", "c"),
    Value = c(1, 2, 3))
```

```r
# Plot with defaults
ggplot(
    data = df, 
    aes(
        x = Name, 
        y = Value)) +
    geom_point()
```

```r
# Plot with parameters
ggplot(
    data = df, 
    aes(
        x = Name, 
        y = Value)) +
    geom_point() +
    ggtitle("Hello World") +
    xlab("Name") +
    ylab("Value")
```

```r
# Create a bar chart
ggplot(
    data = df, 
    aes(
        x = Name, 
        y = Value)) +
    geom_bar(
        stat = "identity", 
        fill = "skyblue") +
    ggtitle("Hello World") +
    xlab("Name") +
    ylab("Value")
```

**2-3 - Qualitative Univariate Analysis**

```r
movies <- read.csv("movie_metadata.csv", na.strings="", stringsAsFactors=FALSE)
movies$content_rating <- as.factor(movies$content_rating)
movies$language <- as.factor(movies$language)
movies$color <- as.factor(movies$color)
movies$gross <- as.numeric(movies$gross)
str(movies)
movies <- movies[movies$content_rating %in% c("G", "Not Rated", "PG", "PG-13", "R", "TV-14"),]
movies <- movies[complete.cases(movies),]
movies$gross <-(movies$gross/1000000)
```

```r
# Create a frequency bar chart for rating
ggplot(
    data = movies, 
    aes(x = content_rating)) + 
    geom_bar() +
    ggtitle("Count of Movies by Rating")
```

```r
xx<- data.frame(Score= 5:100)
xx$Score <- xx$Score *2
xx$Score <- xx$Score /2
xx
```

```r
# Create a horizontal frequency bar chart for genre
ggplot(
    data = movies, 
    aes(x = content_rating)) +
    geom_bar() +
    coord_flip() + # makes it horizontal
    ggtitle("Count of Movies by Rating")
```

```r
# Create cleveland dot plot
ggplot(
    data = movies, 
    aes(x = content_rating)) + 
    geom_point(stat = "count") +
    coord_flip() +
    ggtitle("Count of Movies by Rating")
```

```r
# Create a pie chart
ggplot(
    data = movies, 
    aes(x = "", fill = content_rating)) +
    geom_bar() +
    coord_polar(theta = "y") +
    ggtitle("Count of Movies by Rating") +
    ylab("")
```

```r
# Create a pie chart of awards
movies["HiScore"] <- ifelse(movies$imdb_score>5, TRUE,FALSE)
ggplot(
    data = movies, 
    aes(x = "", fill = HiScore)) +
    geom_bar() +
    coord_polar(theta = "y") +
    ggtitle("Proportion of Movies that Won Awards") +
    ylab("")
```

** Univariate visualizations of a quantitiative variable **

```r
# Create dot plot
ggplot(
    data = movies, 
    aes(x = duration, stat = "count")) +
    geom_dotplot(binwidth = 1) +
    ggtitle("Distribution of Movie Runtimes") +
    xlab("Runtime (minutes)")
```

```r
# Create violin-style dot plot
ggplot(
    data = movies, 
    aes(x = duration, stat = "count")) +
    geom_dotplot(
        binwidth = 1, 
        stackdir = "center") +
    ggtitle("Distribution of Movie Runtimes") +
    xlab("Runtime (minutes)")
```

```r
# Create box plot of runtime
ggplot(
    data = movies, 
    aes(x = duration, y = duration)) +
    geom_boxplot() +
    coord_flip() +
    ggtitle("Distribution of Movie Runtimes") +
    xlab("") +
    ylab("Runtime (minutes)") +
    theme(
        axis.text.y = element_blank(),
        axis.ticks.y = element_blank())
```

```r
# Create histogram with ggplot
ggplot(
    data = movies, 
    aes(x = duration)) +
    geom_histogram(binwidth = 10) +
    ggtitle("Distribution of Movie Runtimes") +
    xlab("Runtime (minutes)")
```

```r
# Create density plot with ggplot
ggplot(
    data = movies, 
    aes(x = duration)) +
    geom_density() +
    ggtitle("Distribution of Movie Runtimes") +
    xlab("Runtime (minutes)")
```

** Bivariate visualizations for two qualitiative variables **

```r
# Create a grouped frequency bar chart
ggplot(
    data = movies, 
    aes(x = content_rating, fill = HiScore)) +
    geom_bar(position = "dodge") +
    ggtitle("Count of Movies by Rating and High Score") +
    scale_fill_discrete(labels = c("No", "Yes"))
```

```r
# Create a stacked frequency bar chart
ggplot(
    data = movies, 
    aes(x = content_rating, fill = HiScore)) +
    geom_bar() +
    ggtitle("Count of Movies by Rating and High Score") +
    scale_fill_discrete(labels = c("No", "Yes"))
```

```r
# Create a 100% stacked frequency bar chart
ggplot(
    data = movies, 
    aes(x = content_rating, fill = HiScore)) +
    geom_bar(position = "fill") +
    ggtitle("Proportion of Movies by Rating and High Score") +
    ylab("Proportion of Movies") +
    scale_fill_discrete(labels = c("No", "Yes"))
```

** Bivariate visualizations for two quantitiative variables **

```r
# Create scatterplot
ggplot(
    data = movies, 
    aes(x = duration, y = gross)) +
    geom_point() + 
    ggtitle("Runtime vs. Box Office Revenue") +
    xlab("Runtime (minutes)") +
    ylab("Box Office ($M)")
```

```r
# Add a linear regression line
ggplot(
    data = movies, 
    aes(x = duration, y = gross)) +
    geom_point() + 
    geom_smooth(method = "lm") +
    ggtitle("Runtime vs. Box Office Revenue") +
    xlab("Runtime (minutes)") +
    ylab("Box Office ($M)")
```

```r
# Create a frequency heatmap
ggplot(
    data = movies, 
    aes(x = duration, y = gross)) +
    stat_bin2d() + 
    ggtitle("Runtime vs. Box Office Revenue") +
    xlab("Runtime (minutes)") +
    ylab("Box Office ($M)")
```

```r
# Create a hexagonal binned frequency heatmap
ggplot(
    data = movies,
    aes(x = duration, y = gross)) +
    stat_binhex() + 
    ggtitle("Runtime vs. Box Office Revenue") +
    xlab("Runtime (minutes)") +
    ylab("Box Office ($M)")
```

```r
# Create a contour plot of density
ggplot(
    data = movies, 
    aes(x = duration, y = gross)) +
    geom_density2d() + 
    ggtitle("Runtime vs. Box Office Revenue") +
    xlab("Runtime (minutes)") +
    ylab("Box Office ($M)")
```

```r
# Create a level plot of density
ggplot(
    data = movies, 
    aes(x = duration, y = gross)) +
    stat_density2d(aes(fill = ..level..), geom = "polygon") + 
    ggtitle("Runtime vs. Box Office Revenue") +
    xlab("Runtime (minutes)") +
    ylab("Box Office ($M)")
```

```r
movies <- movies[complete.cases(movies),]
moviesy <- aggregate(as.numeric(gross) ~ title_year, movies[movies$gross > 0,], sum)
colnames(moviesy)[2] <- "gross"
moviesy
```

```r
# Create a step chart
ggplot(
    data = moviesy, 
    aes(x = title_year, y = gross)) +
    geom_step() +
    expand_limits(y = 0) + 
    ggtitle("Box Office Revenue by Year") +
    xlab("Year") +
    ylab("Box Office ($M)")
```

```r
# Create a line chart
ggplot(
    data = moviesy, 
    aes(x = title_year, y = gross)) +
    geom_line() +
    expand_limits(y = 0) + 
    ggtitle("Box Office Revenue by Year") +
    xlab("Year") +
    ylab("Box Office ($M)")
```

```r
# Create an area chart
ggplot(
    data = moviesy, 
    aes(x = title_year, y = gross)) +
    geom_area() + 
    ggtitle("Box Office Revenue by Year") +
    xlab("Year") +
    ylab("Box Office ($M)")
```

```r
library(dplyr)
```

```r
# Create table of average box office by rating
average <- movies %>%
    select(content_rating, gross) %>%
    group_by(content_rating) %>% 
    summarize(gross = mean(gross)) %>%
    as.data.frame()
average
```

```r
# Create a bivariate bar chart
ggplot(
    data = average, 
    aes(x = content_rating, y = gross)) +
    geom_bar(stat = "identity") +
    ggtitle("Average Box Office Revenue by Rating") +
    xlab("Rating") +
    ylab("Box Office ($M)")
```

```r
# Create a bivariate box plot
ggplot(
    data = movies, 
    aes(x = content_rating, y = gross)) +
    geom_boxplot() +
    ggtitle("Box Office Revenue by Rating") +
    xlab("Rating") +
    ylab("Box Office ($M)")
```

```r
# Create a notched box plot
ggplot(
    data = movies, 
    aes(x = content_rating, y = gross)) +
    geom_boxplot(notch = TRUE) +
    ggtitle("Box Office Revenue by Rating") +
    xlab("Rating") +
    ylab("Box Office ($M)")
```

```r
# Create a violin plot
ggplot(
    data = movies, 
    aes(x = content_rating, y = gross)) +
    geom_violin() +
    ggtitle("Box Office Revenue by Rating") +
    xlab("Rating") +
    ylab("Box Office ($M)")
```
